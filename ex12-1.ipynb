```python
# 주석 달기 하면서 : 이론들 꼼꼼하게 정리하기 + 공식, 패턴
# member_summary.csv : 성별, 지역별, 나이대별 통계가 기록된 파일 만들기

from datetime import *
from time import *

# 함수
def get_birtday(yy, mm, dd) : 
    return date(yy,mm,dd).strftime('%Y-%m-%d') # date() 숫자를 날짜로 조합 , strftime() - 날짜 형식지정 - 문자열

def get_age(yy, mm, dd) : 
    is_birthday_passed = 0 # 생일이 안 지난경우를 찾아서 -1을 저장 
    if datetime.now().month < mm or datetime.now().month == mm and datetime.now().day < dd :
        is_birthday_passed = -1
        
    age = datetime.now().year - yy + is_birthday_passed # 생일 안지난 사람은 나이에서 -1 됨.
    
    return str(age) # 문자열로 return 한다. join하기 위해
    
def get_gender(code) : 
    return '남' if code in [1,3] else '여'

def summary_write_data(data) : 
    for k,v in data.items() : # 딕셔너리의 값을 k, v로 나눠주는 
        summary.write(str(k) + ',' + str(v) +'\n') # write함수는 무조건 str데이터만 취급

def summary_write(gender, region, part, age) : 
    summary.write('성별 요약\n')
    summary_write_data(gender)
    
    summary.write('\n지역별 요약\n')
    summary_write_data(region)

    summary.write('\n부서별 요약\n')
    summary_write_data(part)

    summary.write('\n연령별 요약\n')
    summary_write_data(age)   
        

sum_region = {}
sum_gender = {'남' : 0, '여':0}
sum_age = {}
sum_part = {}

with open('member.csv','rt') as source : 
    with open('member_add_data.csv','wt') as copy : 
        with open('member_summary.csv','wt') as summary : 

            # header 생성
            header = source.readline().rstrip('\n') + ',생일,나이(만),성별\n'
            copy.write(header)
            
            while True :   
                line = source.readline() #csv파일은 한줄씩 읽어오기
                if not line : 
                    break
    
                # 불필요한 데이터 제거 
                line = line.rstrip('\r\n').split(',')
    
                # 데이터분리
                yy = int(line[1][:2]) + ( 1900 if line[1][7] in ['1','2'] else 2000) 
                mm = int(line[1][2:4])
                dd = int(line[1][4:6])
                gender_code = int(line[1][7])
                region = line[4][:2]
                part = line[3]
                                    
                # 값 가져오기
                birth = get_birtday(yy, mm, dd)
                age = get_age(yy, mm, dd)
                gender = get_gender(gender_code)
    
                # 통계
                # --- 지역별
                if region not in sum_region :
                    sum_region[region] = 0 # 지역이 없다면 생성
                sum_region[region] += 1

                # --- 성별
                sum_gender[gender] += 1
                
                # --- 부서별
                if part not in sum_part :
                    sum_part[part] = 0 # 부서가 없다면 생성
                sum_part[part] += 1

                # --- 나이별
                age_range = (int(age) // 10) * 10
                if age_range not in sum_age : 
                    sum_age[age_range] = 0 # 나이대 추가
                sum_age[age_range] += 1
    
                # 데이터 추가
                line.append( birth )
                line.append( age ) 
                line.append( gender )
    
                # 리스트 join - 모든 요소가 str이어야 함.
                line = (',').join(line) + '\n' 
                copy.write(line)

            summary_write(sum_gender, sum_region, sum_part, sum_age)
            
print('생일 데이터 추가')
```
